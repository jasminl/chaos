<h1>Chapter 4. The Sierpinski gasket</h1>

```python
import matplotlib.pyplot as plt
import random

from math import pi
import numpy as np
```

```python
num_points = 100000

# Define triangle vertices
a = (0, 0)
b = (1, 0)
c = (0.5, 0.5 * np.tan(pi/3))
point = [a, b, c]
```

```python
x = [np.random.rand(1,)]
y = [np.random.rand(1,)]

for i in range(num_points):
  idx = random.randint(0, 2)
  dx = (point[idx][0] - x[-1])/2
  dy = (point[idx][1] - y[-1])/2
  x.append(x[-1] + dx)
  y.append(y[-1] + dy)
```

```python
plt.figure()
plt.scatter(x, y, s=1)
```
