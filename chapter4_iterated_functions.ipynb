<h1>Chapter 4. Iterated functions</h1>

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Define top-level functions
def f1(x, y):
  return (x/2, y/2)

def f2(x, y):
  return ((1+x)/2, y/2)

def f3(x, y):
  return (x/2, (1+y)/2)

f = [f1, f2, f3]
```

```python
x = 0.5
y = 0.5
iterations = 10000

# Create trajectory to view attractor
trajectory = []
random.seed(1)
for t in range(iterations):
  fct_id = random.randint(0, 2)  # Randomly select one of the three vertices
  (x, y) = f[fct_id](x, y)
  trajectory.append((x, y))
```

```python
plt.figure()
trajectory_np = np.asarray(trajectory)
plt.scatter(trajectory_np[:,0], trajectory_np[:, 1], s = 1)
plt.axis('off')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```
