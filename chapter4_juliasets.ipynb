<h1>Chapter 4. Julia Sets</h1>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set up map parameters
xlim = [-1.3, 1.3]
ylim = [-1.3, 1.3]

#Constant
a = 0.29
b = 0.54

# Grid resolution
N = 1000

# Max number of iterations per square
iter = 100

# Consider diverge if above this threshold
thresh = 10000
```

```python
def fct(x, y, a, b):
  """ One iteration of the map, see p.167.

  Returns, real and imaginary coordinates separetely.
  """
  return x**2 - y**2 + a, (2*x*y + b)
```

```python
#Create grid of initial points
dx = (xlim[1] - xlim[0]) / N
dy = (ylim[1] - ylim[0]) / N
x0 = np.linspace(xlim[0] + dx/2, xlim[1] - dx/2, N)
y0 = np.linspace(ylim[0] + dy/2, ylim[1] - dy/2, N)
out = np.ones((len(y0), len(x0)))

# Iterate over the center of each square, drop if beyond threshold
row = 0
for xx in x0:
  col = 0
  for yy in y0:
    cx, cy = xx, yy
    for t in np.arange(0, iter):
      cx, cy = fct(cx, cy, a, b)
      if np.sqrt(cx*cx + cy*cy) > thresh:
        out[row, col] = 0
        break
    col += 1
  row += 1
```

```python
plt.imshow(out)
```
