<H1>Chapter 5, Tinkerbell Map</H1>

```python
import matplotlib.pyplot as plt
```

```python
# Some function definitions

c_1 = -0.3
c_2 = -0.6
c_3 = 2
c_4 = 0.5

n_iter = 100

def tkmap(x, y):
  """ Tinkerbell map. """
  return x**2 - y**2 + c_1 * x + c_2 * y, 2 * x * y + c_3 * x + c_4 * y

def tkorbit(x_0, y_0):
  x = [x_0]
  y = [y_0]

  for t in range(0, n_iter):
    xt, yt = tkmap(x[-1], y[-1])
    x.append(xt)
    y.append(yt)
  return x, y
```

<h2>Compute orbit</h2>

```python
x, y = tkorbit(0.1, 0.1)
```

```python
plt.figure()
plt.scatter(x, y, s=1)
plt.show()
```

<h2>Vary $c_4$</h2>

```python
c_4 = 0.1
x, y = tkorbit(0.1, 0.1)
plt.figure()
plt.scatter(x, y, s=1)
plt.show()
```

```python
c_4 = -0.1
x, y = tkorbit(0.1, 0.1)
plt.figure()
plt.scatter(x, y, s=1)
plt.show()
```
